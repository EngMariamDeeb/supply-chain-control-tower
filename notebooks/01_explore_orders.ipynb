```python
orders['order_date'] = pd.to_datetime(orders['order_date'])
orders['expected_delivery_date'] = pd.to_datetime(orders['expected_delivery_date'])
orders['actual_delivery_date'] = pd.to_datetime(orders['actual_delivery_date'])
```

```python
orders.info()
```

```python
orders['delivery_delay_days'] = (orders['actual_delivery_date'] - orders['expected_delivery_date']).dt.days
orders
```

```python
mask = orders['order_status'] == 'Delivered' 
delivered_orders = orders[mask]
# delivered_orders = orders[orders['orders_status'] == 'delivered']
delivered_orders 
```

```python
on_time_mask = delivered_orders['delivery_delay_days'] <= 0
total_delivered = len(delivered_orders)
on_time_rate = (on_time_mask.sum() / total_delivered) * 100
on_time_rate
print(f"On-Time Delivery Rate: {on_time_rate:.1f}%")
```

```python
orders.info()
orders
```

```python
total_ordered = delivered_orders['quantity_ordered'].sum()
total_delivered_units = delivered_orders['quantity_delivered'].sum()
fill_rate = (total_delivered_units / total_ordered) * 100
print(f"Fill Rate: {fill_rate:.1f}%")
```

```python
currently_delayed_count = (orders['order_status'] == 'Delayed').sum()
delivered_late_count = (delivered_orders['delivery_delay_days'] > 0).sum()

print(f"Currently Delayed (in-progress): {currently_delayed_count}")
print(f"Delivered Late (historical): {delivered_late_count}")
```
